### **6. Escala los datos, si consideras necesario**

```python
#6. Escala los datos, si consideras necesario, con StandardScaler() y explica los resultados

#Escala los datos con StandardScaler()
scaler = StandardScaler()

#Ajusta el modelo
scaler.fit(df_numerical)

#Transforma los datos
scaled_data = scaler.transform(df_numerical)

#Convierte los datos a un DataFrame
scaled_data = pd.DataFrame(scaled_data, columns=df_numerical.columns)

#Muestra los datos escalados
scaled_data.head()


```


#### EXPLICACIÓN DE LOS RESULTADOS
>Los datos se han escalado para que tengan una media de 0 y una desviación estándar de 1. Esto es necesario para que los datos no tengan una influencia mayor que otros en el análisis. 
Por ejemplo, si no se escalan los datos, la variable Edad tendrá una influencia mayor que la variable Monto, ya que la variable Edad tiene valores más altos que la variable Monto.
Con los datos escalados, la variable Edad tendrá una influencia similar que la variable Monto, ya que los datos de ambas variables se encuentran en el mismo rango.
Los datos escalados se encuentran en el rango de -3 a 3, ya que la desviación estándar es 1 y la media es 0.


### **7. Reduce las dimensiones con PCA, si consideras necesario.**

```python

#Reduce las dimensiones con PCA
from os import PRIO_PROCESS


pca = PCA(n_components=4)

#Ajusta el modelo
pca_data = pca.fit_transform(scaled_data)


summary = pd.DataFrame({
    "Desviación estándar": np.sqrt(pca.explained_variance_),
    "Varianza proporcional": pca.explained_variance_ratio_,
    "Varianza acumulada": np.cumsum(pca.explained_variance_ratio_)
})

summary = summary.transpose()

summary.columns = ["PC{}".format(i) for i in range(1, len(summary.columns) + 1)]

summary.round(4)
```

```python
pca = PCA(n_components=9)
pca_data = pca.fit_transform(scaled_data)

#Convierte los datos a un DataFrame
summary = pd.DataFrame({
    "Desviación estándar": np.sqrt(pca.explained_variance_),
    "Varianza proporcional": pca.explained_variance_ratio_,
    "Varianza acumulada": np.cumsum(pca.explained_variance_ratio_)
})

summary = summary.transpose()

summary.columns = ["PC{}".format(i) for i in range(1, len(summary.columns) + 1)]

summary.round(4)
```

### **8. Elabora los histogramas de los atributos para visualizar su distribución**

```python
PC_components = summary.columns
PC_components
```

```python
PC_components = summary.columns
#PC_components

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = pca.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components, 
                 y = np.cumsum(pca.explained_variance_ratio_), 
                 color = 'black', 
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Histograma')
plt.xlabel('N-th Componentes principales')
plt.ylabel('Varianza explicada')
plt.ylim(0, 1)
plt.show()
```
