### **1. Descarga los datos a un sitio externo. y carga el dataset en tu libreta.**

```python
import pandas as pd
path = "./default of credit card clients.csv"
df = pd.read_csv(path)
```

### **2. Obten la información del DataFrame con los métodos y propiedades: shape, columns, head(), dtypes, info(), isna()**

```python
#Obten la información del DataFrame con los métodos y propiedades: shape, columns, head(), dtypes, info(), isna()

#Obtenemos la información del DataFrame con el método shape
print("El DataFrame tiene {} filas y {} columnas".format(df.shape[0], df.shape[1]))
```

```python
#Obtenemos la información del DataFrame con el método columns
print("Las columnas del DataFrame son: {}".format(df.columns))
```

```python
#Obtenemos la información del DataFrame con el método head()
print("Las primeras 5 filas del DataFrame son: {}".format(df.head()))
```

```python
#Obtenemos la información del DataFrame con el método dtypes
print("Los tipos de datos del DataFrame son: {}".format(df.dtypes))
```

```python
#Obtenemos la información del DataFrame con el método info()
print("La información del DataFrame es: {}".format(df.info()))
```

```python
#Obtenemos la información del DataFrame con el método isna()
print("El DataFrame tiene {} valores nulos".format(df.isna().sum().sum()))
```

### **3. Limpia los datos eliminando los registros nulos o rellena con la media de la columna**

```python

#Eliminamos los registros nulos
df = df.dropna()

```

```python
#Rellena los valores nulos con la media de la columna
df = df.fillna(df.mean())
```

### **4. Calcula la estadística descriptiva con describe() y explica las medidas de tendencia central y dispersión**

```python
#Calcula la estadística descriptiva con describe() y explica las medidas de tendencia central y dispersión

#Calculamos la estadística descriptiva con el método describe()
print("La estadística descriptiva del DataFrame es: {}".format(df.describe()))
```

```python
#Calcula la correlación entre las variables con corr() y explica los resultados

#Calculamos la correlación entre las variables con el método corr()
print("La correlación entre las variables del DataFrame es: {}".format(df.corr()))
```

```python
#Crea un heatmap con la correlación entre las variables

#Importamos la librería seaborn
import seaborn as sns

#Creamos el heatmap
mapa_calor = df.corr()
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.heatmap(mapa_calor, vmin = -1, vmax = 1, cmap = "BuGn", annot= True)


```

### **5. Realiza el conteo de las variables categóricas**

```python

#Realizamos el conteo de las variables categóricas con el método value_counts()
print("El conteo de las variables categóricas del DataFrame es: {}".format(df.value_counts()))
```

### **6. Escala los datos, si consideras necesario**

```python
#6. Escala los datos, si consideras necesario, con StandardScaler() y explica los resultados

#Importamos la librería StandardScaler
from sklearn.preprocessing import StandardScaler


#Creamos el objeto StandardScaler
scaler = StandardScaler()

#Estandarizamos los datos
dfScaler = scaler.fit_transform(df)

#Creamos el DataFrame con los datos estandarizados
dfFit = pd.DataFrame(dfScaler, columns = df.columns) 
#Mostramos los datos estandarizados
print("Los datos estandarizados son: {}".format(dfFit))
```

### **7. Reduce las dimensiones con PCA, si consideras necesario.**

```python
#7. Reduce las dimensiones con PCA, si consideras necesario.

#Importamos la librería PCA
from sklearn.decomposition import PCA

#Creamos el objeto PCA
pca = PCA()

#Reducimos las dimensiones
dfPCA = pca.fit_transform(df)

#Creamos el DataFrame con los datos reducidos
dfFit = pd.DataFrame(dfPCA, columns = df.columns)
#Mostramos los datos reducidos
print("Los datos reducidos son: {}".format(dfFit))
```

```python
# Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploración de los datos la varianza > 70%

#Calculamos la varianza de los datos explicada por cada componente seleccionado
print("La varianza de los datos explicada por cada componente seleccionado es: {}".format(pca.explained_variance_ratio_))


#Indica la importancia de las variables en cada componente

#Calculamos la importancia de las variables en cada componente
print("La importancia de las variables en cada componente es: {}".format(pca.components_))

```
